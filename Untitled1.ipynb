```python
import numpy as np
import pandas as pd

def calculate_day_idic(solar_flux, regional_baseline_field=48.0):
    geomagnetic_glut = (solar_flux / 100.0) ** 1.8 * (regional_baseline_field * 0.05)
    magneto_short_circuit = 1.0 - np.exp(-0.08 * geomagnetic_glut)
    salmon_bottleneck_density = magneto_short_circuit * 0.85
    base_hunting_hours = 8.0
    actual_hunting_hours = base_hunting_hours * (1.0 - salmon_bottleneck_density)
    cognitive_slack_hours = base_hunting_hours - actual_hunting_hours
    cultural_fad_prob = 1.0 / (1.0 + np.exp(-(cognitive_slack_hours - 4.5)))
    return geomagnetic_glut, magneto_short_circuit, cognitive_slack_hours, cultural_fad_prob

def process_space_weather_csv(file_path):
    df = pd.read_csv(file_path)
    df.columns = df.columns.str.strip()
    gluts, failures, slacks, probabilities = [], [], [], []
    for flux_value in df['Solar_Flux']:
        glut, fail, slack, prob = calculate_day_idic(flux_value)
        gluts.append(round(glut, 2))
        failures.append(round(fail * 100, 1))
        slacks.append(round(slack, 2))
        probabilities.append(round(prob * 100, 1))
    df['Geomagnetic_Glut_uT'] = gluts
    df['Magnetoreception_Failure_Pct'] = failures
    df['Cognitive_Slack_Hours'] = slacks
    df['Cultural_Fad_Probability_Pct'] = probabilities
    df_results = df.sort_values(by='Cultural_Fad_Probability_Pct', ascending=False)
    output_filename = "IDIC_Framework_Results.csv"
    df_results.to_csv(output_filename, index=False)
    return df_results.head(10)

if __name__ == "__main__":
    test_fluxes = [100.0, 150.0, 165.5, 200.0, 250.0, 300.0]
    for flux in test_fluxes:
        glut, fail, slack, prob = calculate_day_idic(flux)
        print(f"{flux},{glut:.2f},{fail*100:.1f},{slack:.2f},{prob*100:.1f}")
```

### Understanding the IDIC Framework Functions

The code provided defines two main functions:

1.  **`calculate_day_idic(solar_flux, regional_baseline_field=48.0)`**:
    This function takes a `solar_flux` value and calculates several derived metrics:
    *   `geomagnetic_glut`: A measure of geomagnetic activity influenced by solar flux.
    *   `magneto_short_circuit`: Represents a disruption, possibly related to magnetoreception.
    *   `salmon_bottleneck_density`: Derived from the short-circuit, implying a density factor for salmon.
    *   `actual_hunting_hours`: Reduced hunting hours due to the bottleneck density.
    *   `cognitive_slack_hours`: The difference between base and actual hunting hours.
    *   `cultural_fad_prob`: A probability derived from cognitive slack hours. This metric, `Cultural_Fad_Probability_Pct`, seems to be the most relevant output for tracking the 'salmon hat phenomenon' in relation to solar flares.

2.  **`process_space_weather_csv(file_path)`**:
    This function reads a CSV file (expected to have a 'Solar_Flux' column) and applies the `calculate_day_idic` function to each solar flux value. It then adds the calculated metrics as new columns to the DataFrame and saves the results to a new CSV file named `IDIC_Framework_Results.csv`. It also returns the top 10 rows of the results DataFrame.

### Visualizing the Correlation

To make it even easier to spot potential correlations with the 'salmon hat phenomenon', let's visualize the `Cultural_Fad_Probability_Pct` over time for 1987 and 2022. This will help you see any peaks where the probability was high. We can then save this plot as an image.
